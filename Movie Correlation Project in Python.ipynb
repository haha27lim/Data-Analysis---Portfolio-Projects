```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

df = pd.read_csv(r"C:\Users\Eileen\Downloads\Python Tutorials\movies.csv")
```

```python
df.head()
```

```python
# We need to see if we have any missing data
# Let's loop through the data and see if there is anything missing

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# there are missing values, fill up with 0
df = df.fillna(0)
```

```python
# Data Types for our columns

print(df.dtypes)
```

```python
# Change data type of columns

df['budget'] = df['budget'].astype('int64')

df ['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
# Create the year from released date
df['year_released'] = df['released'].astype(str).str.extract(r'(\d{4})')

df
```

```python
df = df.sort_values(by=['gross'], inplace= False, ascending= False)
```

```python
pd.set_option('display.max_rows', 20)
```

```python
df.sort_values(by=['gross'], inplace= False, ascending= False)
```

```python
# Drop any duplicates

df['company'].astype(str).drop_duplicates().sort_values(ascending=False)
```

```python
# Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Budget for Film')

plt.ylabel('Gross Earnings')

plt.show()
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x= 'budget', y='gross', data=df, scatter_kws={'color': 'red'}, line_kws={'color': 'blue'})
plt.show()
```

```python
correlation_matrix = df.corr(method= 'pearson', numeric_only = True)

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
       df_numerized[col_name] = df_numerized[col_name].astype('category')
       df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
correlation_matrix = df_numerized.corr(method= 'pearson', numeric_only = True)

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
df_numerized.corr(numeric_only = True)
```

```python
correlation_mat = df_numerized.corr(numeric_only = True)

corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values()

print(sorted_pairs)
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

print(high_corr)
```
